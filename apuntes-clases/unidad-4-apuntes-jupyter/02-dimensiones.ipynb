#### conociendo la biblioteca Matplotlib
el paquete ""Matplotlib"" permite crear visualizuaciones (graficos) para interpretar resultados y data

```python
import matplotlib.pyplot as plt
```

#### creando un grafico simple

```python
plt.plot(vector, label = 'vector')
plt.title('grafico del vector')
plt.xlabel('indice (posición)') # agrega el titulo del eje x
plt.ylabel('valores') # agrega el titulo del eje y

plt.grid(True) # agrega una grilla al gráfico
plt.legend(
    



) # mostrar una etiqueta, que se obtiene de la propiedad label del metodo plot()

plt.show() #permite visualizar el gráfico
```

#### crear dos vectores en el mismo grafico

```python
vector1 = np.random.randint(1, 11, 12)
vector2 = np.random.randint(5, 21, 12)
vector1
vector2
```

```python
plt.plot(vector1, marker = 'o', label = 'vector 1')
plt.plot(vector2, marker = 'o', label = 'vector 2')

plt.title('comparando dos vectores')
plt.xlabel('indice (posición)')
plt.ylabel('valores')

plt.grid(True)
plt.legend()
```

#### Creando un gráfico de barra

```python
plt.bar(range(len(vector2)), vector2) # el metodo bar, permite plotear un grafico de barras
```
