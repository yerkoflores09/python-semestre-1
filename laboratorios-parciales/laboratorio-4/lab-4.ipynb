#### **Importando librerias**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### **Accediendo a archivo csv, iniciando dataset**

```python
pokemon = pd.read_csv('../All_Pokemon.csv', index_col='Number')
pokemon.head()
```

#### **Información general del Dataset, y la información estadística resumida.**

```python
type(pokemon)

print(f'la cantidad de filas y la cantidad de columnas: {pokemon.shape}\n')

pokemon.info()
```

```python
pokemon.describe()
```

#### Filas duplicadas

```python
pokemon_duplicados = pokemon.duplicated()
pokemon_duplicados
```

#### eliminar las filas dulpicadas

```python
pokemon_sin_duplicados = pokemon.dropna()
pokemon_sin_duplicados
```

#### Contar todos los valores nulos existentes en el dataset (Columnas)

```python
columnas_nulas = pokemon.isnull().any()
columnas_nulas
```

#### Contar todos los valores nulos existentes en el dataset (filas)

```python
filas_nulas = pokemon[pokemon.isnull().any(axis=1)]
filas_nulas
```

#### **nuevo dataframe sin datos nulos**

```python
pokemon_sin_nulos = pokemon.dropna()
pokemon_sin_nulos
```

#### **filtrar los pokemons de tipo fuego**

```python
def filtro_fuego(pokemon):
    return 'RPG' in pokemon
```

#### promedio de defensa tipos fuego

```python
promedio_defensa = filtro_fuego['Defense'].mean()
print(f"Promedio defensa (tipo fuego): {promedio_defensa:.2f}")
```
